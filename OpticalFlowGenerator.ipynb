```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
peds1_train_path = './UCSD_Anomaly_Dataset.v1p2/UCSDped1/Train/'
peds2_train_path = './UCSD_Anomaly_Dataset.v1p2/UCSDped2/Train/'
peds1_train_filelist = os.listdir('./UCSD_Anomaly_Dataset.v1p2/UCSDped1/Train/')
peds2_train_filelist = os.listdir('./UCSD_Anomaly_Dataset.v1p2/UCSDped2/Train/')
```

```python
sample_train_path = './UCSD_Anomaly_Dataset.v1p2/UCSDped1/Train/Train001/'
flow_path = './UCSD_Anomaly_Dataset.v1p2/UCSDped1/Train/Flow001/'

filenames = os.listdir(sample_train_path)
filenames = [i for i in filenames if i not in ['.DS_Store', '._.DS_Store']]
```

```python
frame_1 = cv2.imread(sample_train_path + filenames[0])
frame_1 = cv2.resize(frame_1, (512, 512))

hsv = np.zeros_like(frame_1)
hsv[:,:, 1] = 255

frame_1 = cv2.cvtColor(frame_1, cv2.COLOR_BGR2GRAY)


fgbg = cv2.createBackgroundSubtractorMOG2()
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

for i in filenames:
    frame_2 = cv2.imread(sample_train_path + i)
    frame_2 = cv2.cvtColor(frame_2, cv2.COLOR_BGR2GRAY)
    frame_2 = cv2.resize(frame_2, (512, 512))
    
    print(frame_1.shape, frame_2.shape)
    flow = cv2.calcOpticalFlowFarneback(frame_1, frame_2, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    
    frame_1 = frame_2
    
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[:,:, 1])
    hsv[:,:, 0] = ang * 180 / np.pi / 2
    hsv[:,:, 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)

    _, thresh = cv2.threshold(gray, 25, 255, cv2.THRESH_BINARY)
    
    fgmask = fgbg.apply(frame_1)
    fgmask = cv2.morphologyEx(fgmask, cv2.MORPH_OPEN, kernel)
    
    cv2.imwrite(flow_path + str(len(os.listdir(flow_path))) + '.jpg', fgmask)
```
